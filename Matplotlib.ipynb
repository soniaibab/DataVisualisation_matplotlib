# DATASET 1: IRIS 

```python
df=pd.read_csv("Iris.csv")
```

```python
df.head()
```

```python
data=df.drop(['Unnamed: 5','Unnamed: 6', 'sepal length.1','Unnamed: 8'], axis=1)
```

```python
data.head()
```

```python
pyplot.hist(data['petal length'],bins=50,align='mid',histtype='bar', density=True,color="green")
pyplot.xlabel("Length in cms")
pyplot.title("Histogram of Sepal length")
pyplot.grid(True)
pyplot.show()
```

```python
pyplot.scatter(data['sepal length'],data['petal width'], color='green',label='sepal length', alpha=0.5,  marker='o')
pyplot.scatter(data['petal length'],data['petal width'],color='blue', label='petal length', alpha=0.5, marker='^')
pyplot.xlabel('petal_width')
pyplot.ylabel('petal_and_sepal_length')
pyplot.title('Scatter plot between sepal and petal length w.r.t petal_width')
pyplot.legend()
pyplot.show()


```

```python
data.boxplot(column='sepal length', by='iris')
pyplot.show()
```

```python
pyplot.boxplot(data['sepal length'])
pyplot.show()
```

```python
import seaborn as sns
```

```python
sns.violinplot(x=data['iris'], y=data['sepal length']) 
```

```python
width=0.5
x=data['sepal length']
y=data['sepal width']
pyplot.plot(x,y,width,color='g', linestyle='dashed', marker='o', markerfacecolor='blue')
pyplot.show()
pyplot.bar(x,y,width,color='y')
pyplot.show()
```

# DATASET 2: DIABETES

```python
df=pd.read_csv("diabetes_csv.csv")
```

```python
df.head()
```

```python
pyplot.scatter(df['age'],df['pres'], color='green',label='pres', alpha=0.5,  marker='o')
pyplot.scatter(df['age'],df['plas'],color='blue', label='plas', alpha=0.5, marker='^')
pyplot.xlabel('age')
pyplot.ylabel('pres and plas')
pyplot.title('Scatter plot between age and pres and plas')
pyplot.legend()
pyplot.show()

```

```python
pyplot.hist(df['age'],bins=50,align='mid',histtype='bar', density=True,color="green")
pyplot.xlabel("age in years")
pyplot.title("Histogram of Age of people having diabetes")
pyplot.grid(True)
pyplot.show()
```

```python
df.boxplot(column='age', by='class')
pyplot.show()
```

```python
sns.violinplot(x=df['class'], y=df['pres']) 
```

```python
width=0.5
x=df['age']
y=df['pres']
pyplot.plot(x,y,width,color='g',linestyle='dashed', marker='o', markerfacecolor='blue')
pyplot.xlabel('age')
pyplot.ylabel('pres')
pyplot.show()
```
